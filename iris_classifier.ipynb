```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score ,ConfusionMatrixDisplay
from sklearn.tree import DecisionTreeClassifier ,plot_tree
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the iris dataset
iris = load_iris()
# fitting Features and target variable to X and y respectively
X,y = iris.data ,  iris.target
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test , y_train , y_test = train_test_split(X,y , test_size = 0.25 , random_state = 50)
# Training the Decision Tree Classifier
model = DecisionTreeClassifier()
# Fitting the model
model.fit(X_train , y_train)
```

```python
# Predicting the test set results
y_pred =model.predict(X_test)
# Evaluating the model by finding the accuracy
print("Accuracy : " , accuracy_score(y_test , y_pred))
# Displaying the predictions
print("Predictions : " , y_pred)
```

## Visualising the predicted data through confuson matric and throgh descision tree


```python
#this shows the confusion matrix in which we can see how many predictions were correct and how many were wrong by the color intensity
#darker the color more the correct predictions  
#the diagonal elements are the correct predictions
#the non-diagonal elements are the incorrect predictions
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, display_labels=iris.target_names, cmap='Blues')
plt.show()
```

```python
# Visualizing the Decision Tree
# Create a smaller figure (6x4 inches) for the plot.
plt.figure(figsize=(12,8))

# Plotting the decision tree
plot_tree(model, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```
